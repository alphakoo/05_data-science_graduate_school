# 1. examdata1.csv 파일로부터 아래의 정보를 획득하는 파이썬 프로그램을 작성하시오. (한 개의 프로그램으로 구현) (30점)
## 1) Supplier X에서 가장 높은 부품 가격과 가장 낮은 부품 가격을 구함
## 2) Supplier Y의 Invoice Number 별(50-9501, 50-9505) 평균 부품 가격을 구함
## 3) 평균 부품 가격이 가장 높은 Supplier를 찾음

```python
import csv
import math

product_X = 'Supplier X'
product_Y = 'Supplier Y'
product_Z = 'Supplier Z'

cost_x_list = []
cost_y_num1_list = []
cost_y_num2_list = []
cost_z_list = []

invoice_y_num1 = '50-9501'
invoice_y_num2 = '50-9505'

with open(examdata1_file, 'r') as csv_in_file:

    file_reader = csv.reader(csv_in_file, delimiter=',')
    header = next(file_reader)         

    # print(header)

    for row in file_reader:

        cost_val = float(row[3].strip('$').replace(',', ''))

        if row[0] == product_X:

            cost_x_list.append(cost_val)

        elif row[0] == product_Y:

            if row[1] == invoice_y_num1:

                cost_y_num1_list.append(cost_val)

            elif row[1] == invoice_y_num2:

                cost_y_num2_list.append(cost_val)

        elif row[0] == product_Z:

            cost_z_list.append(cost_val)

        

    # print(f'cost_x_list: {cost_x_list}')
    # print(f'cost_y_num1_list: {cost_y_num1_list}')
    # print(f'cost_y_num2_list: {cost_y_num2_list}')
    # print(f'cost_z_list: {cost_z_list}')
    cost_x_list.sort()
    # print(cost_x_list)

    # 1) solve
    print('------------------------------------------')

    print(f'Highest Cost of Supplier X: {cost_x_list[-1]}')

    print(f'Lowest Cost of Supplier X: {cost_x_list[0]}')
    

    # 2) solve
    ave_num1 = sum(cost_y_num1_list) / len(cost_y_num1_list)
    
    ave_num2 = sum(cost_y_num2_list) / len(cost_y_num2_list)

    print('------------------------------------------')

    print(f'Average Cost of Supplier Y_50-9501: {ave_num1}')

    print(f'Average Cost of Supplier Y_50-9505: {ave_num2}')

    # 3) solve
    ave_cost_X = sum(cost_x_list) / len(cost_x_list)

    ave_cost_Y = (sum(cost_y_num1_list) + sum(cost_y_num2_list)) / (len(cost_y_num1_list) + len(cost_y_num2_list))

    ave_cost_Z = sum(cost_z_list) / len(cost_z_list)

    print('------------------------------------------')

    print(f'Average Cost of Supplier X: {ave_cost_X}')
    print(f'Average Cost of Supplier Y: {ave_cost_Y}')
    print(f'Average Cost of Supplier Z: {ave_cost_Z}')
```

# 3. examdata3.xls 데이터셋에서 아래의 정보를 획득하는 프로그램을 작성하시오. (20점) (한 개의 프로그램으로 구현)
## 1) examdata3.xls파일에 있는 모든 워크시트들에서 Customer ID가 5000이하이고 Sale Amount가 $2,000이하인 행만 필터링함
## 2) examdata3.xls파일에 있는 모든 워크시트들의 상품 판매에 대한 평균 가격을 출력함

```python
import sys
from datetime import date
from xlrd import open_workbook, xldate_as_tuple
from xlwt import Workbook

input_file = '/content/drive/MyDrive/[00]데이터사이언스 대학원/22년 1학기 데이터과학 프로그래밍/기말고사/examdata3.xls'

output_file = '/content/drive/MyDrive/[00]데이터사이언스 대학원/22년 1학기 데이터과학 프로그래밍/기말고사/examsol3.xls'

output_workbook = Workbook()
output_worksheet = output_workbook.add_sheet('filtered')

first_worksheet = True

sale_amount_column_index = 3
customer_id_column_index = 0

sale_amount_list = []

with open_workbook(input_file) as workbook:

    data = []

    for worksheet in workbook.sheets():
        
        worksheet.row_values(0)

        for row_index in range(1, worksheet.nrows):

            row_list = []

            sale_amount = worksheet.cell_value(row_index, sale_amount_column_index)

            customer_id = worksheet.cell_value(row_index, customer_id_column_index)

            sale_amount_list.append(sale_amount)

            if customer_id < 5000 and sale_amount < 2000:

                for column_index in range(worksheet.ncols):

                    cell_value = worksheet.cell_value(row_index, column_index)
                    cell_type = worksheet.cell_type(row_index, column_index)

                    if cell_type == 3:

                        date_cell = xldate_as_tuple(cell_value, workbook.datemode)
                        date_cell = date(*date_cell[0:3]).strftime('%m/%d/%Y')

                        row_list.append(date_cell)

                    else:

                        row_list.append(cell_value)

                if row_list:

                    data.append(row_list)

    
    for list_index, output_list in enumerate(data):

        for element_index, element in enumerate(output_list):

            output_worksheet.write(list_index, element_index, element)


output_workbook.save(output_file)     


# 2) solve
ave_sale_amount = sum(sale_amount_list) / len(sale_amount_list)

print(f'Average of Sale_amount: {round(ave_sale_amount, 2)}')
```

## 2) Vehicle 클래스를 상속받는 ElectricVehicle 클래스를 정의하고, "Fully charged"라는 메시지를 출력할 수 있는 charge 메서드를 추가 구현한 후, ElectricVehicle 클래스에 대한 객체를 생성한 후, charge 메서드를 실행하는 프로그램을 작성하시오.
## 3) Vehicle 클래스를 상속받는 ElectricVehicle 클래스에서 drive 메서드를 overriding하여 "A car is stopped"가 출력되도록 하시오.

```python
class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def drive(self):
        print(' A car is driving fast.')
```

```python
# 1)solve
my_car = Vehicle('Hyundai', 'Sonata')

print(f'my_car\'s brand: {my_car.brand}')
print(f'my_car\s model: {my_car.model}')
```

```python
# 2)solve
class ElectricVehicle(Vehicle):

    def charge(self):
        print('Fully Charged!')

    # overriding
    def drive(self):
        print(' A car is stopped.')        

my_elec_car = ElectricVehicle('Kia', 'Morning')

my_elec_car.charge()
```

```python
# 3)solve
my_elec_car.drive()
```
